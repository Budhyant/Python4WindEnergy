The read_csv function has an handy date_parser argument that allows the user to define her/his own date parse function. The default one is:

````python
import dateutil.parser as parser
def date_parser_default(date):
    date = parser.parse(date)
    return date
````

Because the parser tries to actually understand the date format it is very slow. If we specify how to read the date field, it can go a lot faster. As a bonus, we could include any other correction or transformation operation on the date.

````python
from datetime import datetime
def date_parser(date):
    # create a datetime instance and assume constant formatting
    # format: 20131211 18:29:30
    return datetime(int(date[:4]), int(date[4:6]), int(date[6:8]), int(date[9:11]), int(date[12:14]), int(date[15:17]))
df = pd.read_csv(fname, header=0, skiprows=[1], parse_dates=[0], date_parser=date_parser)
````

````python
%timeit pd.read_csv(fname, header=0, skiprows=[1], parse_dates=[0], date_parser=date_parser)
````

Looks a bit ugly, isn't the next one more Pythonic?

````python
def date_parser_2(date):
    return datetime.strptime(date, "%Y%m%d %H:%M:%S")
%timeit pd.read_csv(fname, header=0, skiprows=[1], parse_dates=[0], date_parser=date_parser_2)
````

Is the above histogram presenting a fair comparison between the 3 months? If not, can you use simple pandas methods to reduce the bias that possibly one month has over another...

````python
#The above histogram is not presenting a fair comparison between the 3 months, since the number of 
#samples differ quite substantially. In addition, the measurement campaign was performed in a period
#of time where datasets don't overlap. This will limit any attempt to completely suppress the bias 
#between moths (without making crude assumptions). However, it is possible to reduce the bias
#as shown next:

#0) Import libraries and define useful functions.
import numpy as np
import pandas as pd
import csv
from matplotlib import pyplot as plt
from datetime import datetime

#Definition of function "date_parser" that parse date type objects when reading data from CSV.
def date_parser(date):
    # create a datetime instance and assume constant format: 20131211 18:29:30
    return datetime(int(date[:4]), int(date[4:6]), int(date[6:8]), int(date[9:11]), int(date[12:14]), int(date[15:17]))


#1) Import the CSV data (skip row 1 (2 in CSV) containing data units).
df = pd.read_csv(fname, header=0, skiprows=[1], parse_dates=[0], date_parser=date_parser)
print "Length of Original CSV File (df): " + str(len(df))

#2) Dissect information: remove bad measurements by applying selection criteria on the data set.
df_dis1 = df[df['Latitude N'] > 0.0]
df_dis2 = df_dis1[df_dis1['Longitude E'] > 0.0]
df_dis3 = df_dis2[(df_dis2['orientation'] < 360.0) & (df_dis2['orientation'] > 0.0)]
print "Length of Dissected Data (df_dis3): " + str(len(df_dis3))

#3) Indexing: Index data by using the Time column in order to use more advanced selection, interpolation and group.
df_dis3_ind = df_dis3.set_index('Time')
print "Length of Dissected and Indexed Data (df_dis3_ind): " + str(len(df_dis3_ind))

#4) Re-sample the data set in a 5-minutely averaged frequency and fill in the gaps by interpolating the data.
df_resample = df_dis3_ind.resample('5T',how='mean',fill_method='pad')
print "Length of Resampled Data (df_resample): " + str(len(df_resample))

#5) Reindex and re-sample again:
df_resample = df_resample.reindex(pd.date_range('2013-11-14 06:05:00','2014-01-08 16:40:00',freq='5T'))
df_resample = df_resample.resample('5T',how='mean',fill_method='pad')
print "Length of Reindexed Data (df_resample): " + str(len(df_resample))
print df_resample

#Finally, a histogram showing the re-sampled and interpolated data is shown:
df_grouped = df_resample.groupby(lambda x: x.month)
for name, group in df_grouped:
    print ('%3i %6i %6.1f %6.1f') % (name, group.orientation.count(), group.orientation.min(), group.orientation.max())
    ax = group.orientation.hist(alpha=0.6, label=str(name), bins=range(0,361,20))
ax.legend(loc='best')
ax.set_xlim([0,360])


#The difference between the number of samples (among the considered months) has been reduced considerably. 
#However, bias is still too high. The ultimate solution will be to obtain measurements in concurrent periods 
#of time in order to perform fair comparisons.

````
