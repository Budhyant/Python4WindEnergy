If you like linux command, you can use grep to catch all the lines with an email sign

```python
%%bash
echo "---- List of emails -----"
grep @ emails.txt
```

You can import the bash command outputs directly as python variables using the ! command

```python
emails = !grep @ emails.txt
print emails
```

Now let's keep only the usernames (pure python)

```python
user_names = [e.split('@')[0] for e in emails]
print user_names
```
