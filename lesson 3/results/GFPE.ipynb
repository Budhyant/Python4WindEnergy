```python
### Countour Plotting

import math
from pylab import *

alpha = 0.7
phi_ext = 2 * math.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*math.pi, 100)
phi_p = np.linspace(0, 2*math.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T


fig, ax = plt.subplots()

cnt = contour(Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])



# Second Axis
ax2 = ax.twinx()

ax2.set_ylabel(r"Temperature ($^\circ$C)")
ax2.set_ylim(0, 35)



#Add Line
axvline(x=0.5, ymin=0, ymax=1)
axhline(y=25,xmin=0,xmax=1)


#autoformat dates for nice printing on the x-axis using fig.autofmt_xdate()
fig.autofmt_xdate(bottom=0.2, rotation=30, ha='right')
```

```python

x= np.linspace(0, 5*math.pi, 1000)

Y= np.sin(x)
noise = np.random.normal(-0.1,0.1,1000)

y=Y+noise

fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(8,6))
fig.suptitle("Sin + Noise", fontsize=16)
ax1.plot(x, y,color="black", label="Sin(x) + Noise")

z2 = np.polyfit(x,y,2)
z4 = np.polyfit(x,y,4)
z6 = np.polyfit(x,y,6)
z10 = np.polyfit(x,y,10)

p2 = np.poly1d(z2)
p4 = np.poly1d(z4)
p6 = np.poly1d(z6)
p10 = np.poly1d(z10)


plt.plot(x,p2(x),color="red", linewidth=2.00,label="Fit order:2")
plt.plot(x,p4(x),color="green", linewidth=2.00,label="Fit order:4")
plt.plot(x,p6(x),color="purple", linewidth=2.00,label="Fit order:6")
plt.plot(x,p10(x),color="blue", linewidth=2.00,label="Fit order:10")

ax1.legend(fontsize=12, loc="best")




```
